```python
import re
df = pd.read_csv('healthcare_dataset.csv')


# Function to convert column names to camel case
def to_camel_case(name):
    name = re.sub(r'(?<!^)(?=[A-Z])', ' ', name).title().replace(" ", "")
    return name[0].lower() + name[1:]

# Rename all column names to camel case
df.columns = [to_camel_case(col) for col in df.columns]

# Function to convert name values to camel case
def to_camel_case_name(name):
    components = name.split(' ')
    return components[0].lower() + ''.join(x.title() for x in components[1:])

# Apply the function to the 'name' column
df['name'] = df['name'].apply(to_camel_case_name)

# Print the updated DataFrame
print(df)

# Save the DataFrame to a new CSV file
df.to_csv('healthcare_dataset.csv', index=False)
```

```python
import pandas as pd

# Sample DataFrame
data = {
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [30, 45, 25],
    'gender': ['F', 'M', 'M'],
    'bloodType': ['A+', 'B-', 'O+'],
    'medicalCondition': ['Diabetes', 'Hypertension', 'Healthy'],
    'dateOfAdmission': ['2023-01-01', '2023-01-02', '2023-01-03'],
    'doctor': ['Dr. Smith', 'Dr. Jones', 'Dr. Brown'],
    'hospital': ['General Hospital', 'City Hospital', 'County Hospital'],
    'insuranceProvider': ['Insurance A', 'Insurance B', 'Insurance C'],
    'billingAmount': [1000, 1500, 2000],
    'roomNumber': [101, 102, 103],
    'admissionType': ['Emergency', 'Scheduled', 'Emergency'],
    'dischargeDate': ['2023-01-10', '2023-01-15', '2023-01-05'],
    'medication': ['Med A', 'Med B', 'Med C'],
    'testResults': ['Positive', 'Negative', 'Positive']
}

df = pd.DataFrame(data)

# Create Dimension Tables
patient_dimension = df[['name', 'age', 'gender', 'bloodType', 'medicalCondition']].drop_duplicates().reset_index(drop=True)
patient_dimension['patient_id'] = patient_dimension.index + 1

doctor_dimension = df[['doctor']].drop_duplicates().reset_index(drop=True)
doctor_dimension['doctor_id'] = doctor_dimension.index + 1

hospital_dimension = df[['hospital']].drop_duplicates().reset_index(drop=True)
hospital_dimension['hospital_id'] = hospital_dimension.index + 1

insurance_provider_dimension = df[['insuranceProvider']].drop_duplicates().reset_index(drop=True)
insurance_provider_dimension['insurance_provider_id'] = insurance_provider_dimension.index + 1

# Display Dimension Tables
print("Patient Dimension:")
print(patient_dimension)
print("\nDoctor Dimension:")
print(doctor_dimension)
print("\nHospital Dimension:")
print(hospital_dimension)
print("\nInsurance Provider Dimension:")
print(insurance_provider_dimension)
```
